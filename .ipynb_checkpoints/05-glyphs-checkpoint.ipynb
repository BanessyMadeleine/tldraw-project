![](images/z-skeleton-only.jpg)

The curve that you have plotted only represents a "skeleton" of the glyph but not its thickness. 
The thickness itself is computed from an evaluation of the speed at which the pen travels (the faster the pen is, the smaller the curve is ; this is how the pen pressure is simulated).

```{exercise} Glyph Collection

 2. Draw all these glyph variants in matplotib ; translate them beforehand so that they have the same center of mass and are (mostly but not entirely) superimposed.
```

![](images/glyphs-superimposed.jpg)

````python
import numpy as np


with open('docs/glyphs.tldr', mode='rt', encoding='utf-8') as text:
    content = text.read()

data = json.loads(content)
#print(alpha['records'])

draw_records = []
for record in data["records"]:
    if record["typeName"] == "shape" and  record["type"] == "draw":
        draw_records.append(record)

#print(len(draw_record))
    
coordinates = []
#Drawing
for i in range(len(draw_records)):
    draw_record = draw_records[i]
    x_0, y_0 = draw_record['x'], draw_record['y']
    points = draw_record['props']['segments'][0]['points']
    x_list, y_list = np.array([x_0 + point['x'] for point in points], dtype='float64'), np.array([-(y_0 + point['y']) for point in points],dtype='float64')

    #rearranging center of mass
    x_list -= np.mean(x_list)
    y_list -= np.mean(y_list)
    coordinates.append((x_list,y_list))
    
    point_list = [(x_list[i],y_list[i]) for i in range(len(x_list))]
    stroke = perfect_freehand.get_stroke(point_list) 
    outline_x, outline_y = [tuple[0] for tuple in stroke], [tuple[1] for tuple in stroke]
    
    plt.plot(x_list, y_list)
    #uncomment to get oulines
    #plt.plot(outline_x, outline_y)
plt.title('Same drawing different attempts')
plt.show()


````

Each glyph may have a different number of samples, so its hard to compare them directly. Additionally they may be a bit rough and could probably use a smoother and denser representation (with many more points).

```{exercise} Interpolation and Resampling
 1. Associate to any glyph represented by its coordinate a path $t \in [0, 1] \mapsto \mathbb{R}^2$ (as a Python function), using some "smooth" interpolation.
 
    **Hint:** use [`scipy.interpolate.interp1d`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html).
 
 2. Sample it uniformly to get exactly 1000 points.

```



````python
from scipy.interpolate import interp1d

for x_list, y_list in coordinates:
    # Remove duplicate x values
    unique_x_values = {}
    for x, y in zip(x_list, y_list):
        unique_x_values[x] = y

    x_list = list(unique_x_values.keys())
    y_list = list(unique_x_values.values())
    
    # Interpolation and sampling
    interpolation_function = interp1d(x_list, y_list, kind='cubic')
    x_smooth = np.linspace(min(x_list), max(x_list), 1000)
    y_smooth = interpolation_function(x_smooth)

````

```{exercise} Glyph Similarity

 1. Come up with a scalar measure of the "dissimilarity" between two resampled glyphs (2 arrays of 1000 points), which is equal to 0 when both glyphs are identical and gets larger as the glyphs become more different.

 2. Compute the dissimilarity between each glyph and the mean glyph and compute the maximal dissimilarity for our collection of small variants.
```

````python
#taking the norm of the difference
def compute_dissimilarity(coordinates_1, coordinates_2):
    min_length = min(len(coordinates_1[0]),len(coordinates_2[0]))
    return np.sum(np.abs(np.array(coordinates_1[0][:min_length])-np.array(coordinates_2[0][:min_length]))+np.abs(np.array(coordinates_1[1][:min_length])-np.array(coordinates_2[1][:min_length])))

mean_coordinates = [mean_x, mean_y]


for i in range(len(coordinates)):
    print(f"Mean and {i}:", compute_dissimilarity(coordinates[i], mean_coordinates))
    
for i in range(len(coordinates)):
    for j in range(i+1, len(coordinates)):
        print(f"{i} and {j}: ",compute_dissimilarity(coordinates[i],coordinates[j]))
````

````python
with open('docs/glyphs-misc.tldr', mode='rt', encoding='utf-8') as text:
    content = text.read()

data = json.loads(content)

draw_records = []
for record in data["records"]:
    if record["typeName"] == "shape" and  record["type"] == "draw":
        draw_records.append(record)

    
coordinates = []
for i in range(len(draw_records)):
    draw_record = draw_records[i]
    x_0, y_0 = draw_record['x'], draw_record['y']
    points = draw_record['props']['segments'][0]['points']
    x_list, y_list = np.array([x_0 + point['x'] for point in points], dtype='float64'), np.array([-(y_0 + point['y']) for point in points],dtype='float64')
    coordinates.append((x_list,y_list))


for i in range(len(coordinates)-1):
    #print(compute_dissimilarity(coordinates[0],coordinates[i+1]))
    if compute_dissimilarity(coordinates[0],coordinates[i+1]) > 10000:
        draw_records[i]['props']['color'] = 'red'
        #print(draw_records[i]['props']['color'])
    else :
        draw_records[i]['props']['color'] = 'green'
        #print(draw_records[i]['props']['color'])

        
print(draw_records[4]['props']['color'])                   
with open('docs/glyphs-classified.tldr', 'w') as file:
    json.dump(data, file)


````
